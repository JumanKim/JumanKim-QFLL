```python
%matplotlib inline
import numpy as np
import lmfit
import os
from matplotlib import pyplot as plt
from lmfit import Model

def lorentz(x,amp,center,FWHM,offset):
    params=[amp,center,FWHM,offset]
    return params[0]/np.pi*(params[2]/2)/((x-params[1])**2+(params[2]/2)**2)+params[3]

gmodel=Model(lorentz)

fig,ax=plt.subplots()
par1=ax.twinx()
data=open(os.getcwd()+"\\1235_to_1245_horizontal_sweep_300_try.txt").read().split()
data=list(map(float,data))
xdata=np.array([data[i] for i in range(0,len(data),3)])
ydata=np.array([data[i] for i in range(1,len(data),3)])
xaxis=np.linspace(np.min(xdata),np.max(xdata),1000)
result=gmodel.fit(ydata,x=xdata,amp=10.,center=1240.,FWHM=10,offset=0)
err=np.array([data[i] for i in range(2,len(data),3)])
pl1=ax.errorbar(xdata,ydata,err,label='exp')
yaxis=lorentz(xaxis,result.params.valuesdict()['amp'],result.params.valuesdict()['center'],result.params.valuesdict()['FWHM'],result.params.valuesdict()['offset'])
pl2=par1.plot(xaxis,yaxis,'r-',label='fit')
ax.set_title("Half-illuminated")
ax.set_xlabel("freq (kHz)")
ax.set_ylabel("Intensity(A.U)")
plt.show()
```

```python
with open(os.getcwd()+"\\300_try_fit.txt",'w') as fp, open(os.getcwd()+"\\300_try_fit_report.txt",'w') as fp2:
    for i in range(len(xaxis)):
        fp.write("%f %f\n" % (xaxis[i],yaxis[i]))
    fp2.write("%f %f %f %f" % (result.params.valuesdict()['amp'],result.params.valuesdict()['center'],result.params.valuesdict()['FWHM'],result.params.valuesdict()['offset']))
```
